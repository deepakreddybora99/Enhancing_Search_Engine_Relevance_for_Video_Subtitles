```python
import sqlite3
import pandas as pd

colus=[]
path=r"C:\Users\hp\Desktop\internship\search engine project\data\eng_subtitles_database.db"
# Connect to the database (replace 'example.db' with your .db file)
conn = sqlite3.connect(path)

# Create a cursor object
cur = conn.cursor()

# Execute a query to select table_name
cur.execute('SELECT name FROM sqlite_master where type="table"')
# Fetch data
table_name = cur.fetchall()
table_name=table_name[0][0]


cur.execute(f"PRAGMA table_info({table_name})")
cols = cur.fetchall()

for col in cols:
  colus.append(col[1])

cur.execute(f"SELECT COUNT(*) FROM {table_name}")

# Fetch the result
length = cur.fetchone()[0]


# Execute a query to select data
cur.execute(f"SELECT * FROM {table_name}")


# Fetch data
data = cur.fetchall()
#creating the
data=pd.DataFrame(data,columns=colus)

# Print the descrption of Data
length=f"Length of {table_name} table: {length}"
table_name=f"Table Name in Database:{table_name}"
columns=f"Column Names in Table :{colus}"



cur.close()
conn.close()
print(table_name, columns, length,sep='\n')
data.head()
```

```python
import zipfile
import io
from tqdm import tqdm, tqdm_notebook, tqdm_pandas
tqdm.pandas()
count=0
def decode_content(binary_data):
  global count
  count=+1
  # Decompress the binary data using the zipfile module
  with io.BytesIO(binary_data) as f:
    with zipfile.ZipFile(f, 'r') as zip_file:
        # Reading only one file in the ZIP archive
        subtitle_content = zip_file.read(zip_file.namelist()[0])

# Now 'subtitle_content' should contain the extracted subtitle content and assuming the content is latin-1 encoded text
    return subtitle_content.decode('latin-1')
df['decoded_content']=df['content'].progress_apply(decode_content)
```

```python
import random
import nltk
nltk.download('stopwords')
import re
from tqdm import tqdm, tqdm_notebook, tqdm_pandas
tqdm.pandas()
sentences_to_remove = [
    "Watch any video online with Open SUBTITLES Free Browser extension osdb link ext",
    "Advertise your product or brand here contact www OpenSubtitles org today",
    "Subtitles by MemoryOnSmells Support us and become VIP member nto remove all ads from",
    "Subtitles by",
    "api.opensubtitles.org"
]
def preprocess(subtitle_data):
  
  #removing unncessary data using regex
  #removing timestamp
  cleaned_text=[re.sub(r'\d{2}:\d{2}:\d{2},\d{3}-->\d{2}:\d{2}:\d{2},\d{3}', '', text) for text in subtitle_data]
   #removing tags
  cleaned_text=re.sub(r'<.*?>','', subtitle_data)
  cleaned_text=re.sub(r'\r\n','',cleaned_text)
  #removing links
  cleaned_text=re.sub(r'(https?://\S+|www\.\S+)','', cleaned_text)
  #removing unwanted sentence
  pattern = '|'.join(re.escape(sentence) for sentence in sentences_to_remove)
  cleaned_text = re.sub(pattern, '', cleaned_text)
  cleaned_text=cleaned_text.strip()
  #converting to lower text
  cleaned_text = cleaned_text.lower()
  #removing special characters
  cleaned_text = re.sub(r'[^\w\s.]', '', cleaned_text)
  cleaned_text=re.sub(r'\s+', ' ', cleaned_text)
  cleaned_text= re.sub(r'\d+', '', cleaned_text)
  cleaned_text=re.sub(r'ï','',cleaned_text)
  cleaned_text=re.sub(r'âª','',cleaned_text)

  return cleaned_text
```

```python
df['subtitles']=df['decoded_content'].progress_apply(preprocess)
df.head()
```

```python
def preprocess_title(data):
    #removing unnecessary characters in text
    data= re.sub(r'\.eng\.1cd','',data).strip()
    data = re.sub(r'episode\s\d+\s\d+','',data).strip()
    #lower the text
    data = data.lower()
    #removing special characters
    data=re.sub(r'[^\w\s.]', '', data)
    data=data.replace('.',' ')
    return data
df['title']=df['name'].progress_apply(preprocess_title)
```
